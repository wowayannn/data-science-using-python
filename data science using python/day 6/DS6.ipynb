```python
class ATM:
    def __init__(self):
        print('object created successfully')
        self.pin = 1111 #public variable
        self.balance = 2000 #public variable

    def change_pin(self):#public method
        self.pin = input('Enter New Pin: ')
        print('YOUR PIN HAS BEEN CHANGED!!')

    def show_balance(self):
        print(f'''YOUR CURRENT BALANCE IS Rs{self.balance}''')

    def atm_performance_status(self):
        print('ATM Working Fine 100%')
```

```python
icic = ATM()
```

```python
icic.show_balance()
```

```python
icic.balance
```

```python
class ATM_version1:
    def __init__(self):
        print('object created successfully')
        self._pin = 1111 #hide
        self._balance = 2000 #hide

    def change_pin(self):#public method
        self.pin = input('Enter New Pin: ')
        print('YOUR PIN HAS BEEN CHANGED!!')

    def show_balance(self):
        print(f'''YOUR CURRENT BALANCE IS Rs{self.balance}''')

    def atm_performance_status(self):
        print('ATM Working Fine 100%')
```

```python
axis = ATM_version1()
```

```python
class ATM_version_2:
    def __init__(self):
        print('object created successfully')
        self.__pin = 1111 #public variable
        self.__balance = 2000 #private variable

    def change_pin(self):#public method
        self.__pin = input('Enter New Pin: ')
        print('YOUR PIN HAS BEEN CHANGED!!')

    def show_balance(self):
        print(f'''YOUR CURRENT BALANCE IS Rs{self.__balance}''')

    def __atm_performance_status(self): #private method
        print('ATM Working Fine 100%')
```

```python
hdfc = ATM_version_2()
```

```python
hdfc._ATM_version_2__balance  #calling private variable
```
